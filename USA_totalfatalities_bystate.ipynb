```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

#connect to the sqlite database
conn = sqlite3.connect('C:/Users/Dhanasree/Code/SavvyCoders/Capstone/dhana-capstone.db')

#Execute a query to retrieve data
query = 'select * from vehicletype_deaths_byGender'
Genderdf = pd.read_sql_query(query , conn)

# close the and connection    
conn.close()

#insert a new column with values
new_column = [23814,23551,22697,22215,23824,26625]
Genderdf['TotalVehicletypeFatalities'] = new_column

male_column = pd.to_numeric(Genderdf['Total_MaleFatalitiesByVehicletype'])
female_column = pd.to_numeric(Genderdf['Total_FemaleFatalitiesByVehicletype'])
i =0
j = [0,1,2,3,4,5]
for val in j:
    male_column[i] = ((male_column[i] / new_column[i]) * 100)
    female_column[i] = ((female_column[i] / new_column[i]) * 100)
    i = i + 1
   
Genderdf['MaleFatalitiesPercent'] = male_column
Genderdf['FemaleFatalitiesPercent'] = female_column
Genderdf = Genderdf.round(2)

#import into CSV file
# Genderdf.to_csv('C:/Users/Dhanasree/Code/SavvyCoders/Capstone/Data-Sets/CleanedGenderFatalities.csv')

y = Genderdf['Year']
x = Genderdf['FemaleFatalitiesPercent']
#create a line graph
plt.plot(x , y)

#set the labels for x-axis and y-axis

plt.xlabel('X-axis')
plt.ylabel('Y-axis')

#set the title of the graph
plt.title('Female fatalities')

#show the graph
plt.show()


    

```
